```python
data_Cabin_test=data.loc[data.Cabin=="nan"]
data_Cabin_train=data.loc[data.Cabin!="nan"]
data_Cabin_train["Cabin_floor"]=data_Cabin_train.Cabin.str.get(0)
def Cabin_no(s):
    l=s.split(" ")
    l=[i[1:] for i in l]
    try:
        l=[int(i) for i in l if i!=""]
        n=round(sum(l)/len(l))
        return n
    except:
        n=np.nan
        return n
temp=[]
for i in data_Cabin_train["Cabin"]:
    temp.append(Cabin_no(i))
data_Cabin_train["Cabin_no"]=temp
data_Cabin_train["Cabin_no"]=data_Cabin_train.Cabin_no.fillna(round(data_Cabin_train.Cabin_no.mean(skipna=True)))
d={'A':0, 'B':1, 'C':2, 'D':3, 'E':4, 'F':5, 'G':6, 'T':7}
data_Cabin_train["Cabin_floor"]=data_Cabin_train.Cabin_floor.replace(d)
del data_Cabin_train["Cabin"],data_Cabin_test["Cabin"]
data_Cabin_train.head()
```

```python
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.model_selection import GridSearchCV
```

```python
gs=GridSearchCV(cv=10,error_score="f1",estimator=LogisticRegression(),param_grid=[{
    "C":[0.5,1,1.5,2],"penalty":["l1","l2"]
}])
gs.fit(data_Cabin_train.drop(["Cabin_floor","Cabin_no"],axis=1),data_Cabin_train["Cabin_floor"])
```

```python
data_Cabin_test["Cabin_floor"]=gs.predict(data_Cabin_test)
data_Cabin_test.head()
```
