```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def relu(X):
    return np.abs(np.multiply((X>0).astype(int),X))
```

```python
A=np.matrix([
    [0,1,0,0],
    [0,0,1,1],
    [0,1,0,0],
    [1,0,1,0]
],dtype=float)
A
```

```python
G=nx.DiGraph(A)
G=nx.relabel_nodes(G,{i:chr(i+97) for i in sorted(G)})
nx.draw(G,labels={i:i for i in sorted(G)},with_label=True)
```

```python
X=np.matrix([
    [i,-i] for i in range(A.shape[0])
],dtype=float)
X
```

```python
D=np.multiply(np.eye(A.shape[0]),A.sum(axis=0))
A_hat=A+np.matrix(np.eye(A.shape[0]))
D_hat=np.multiply(np.eye(A_hat.shape[0]),A_hat.sum(axis=0))
D_hat
```

```python
W = np.matrix([[1, -1, 1],
             [-1, 1, -1]])
relu(D_hat**-1*A_hat*X*W)
```
