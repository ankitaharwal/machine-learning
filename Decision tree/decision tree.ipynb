# Using ID3

```python
class decision_tree:
    """This is implementaion of decision tree"""
    def __init__(self,impurity="gini",max_depth=None):
        self.impurity=impurity
        self.max_depth=max_depth
        self.model=None
    def entropy(self,*args):
        s=0
        sum_args=sum(args)
        for arg in args:
            if arg==sum_args:
                return 0
            p=arg/sum_args
            if p==0:
                continue
            s+=-p*(np.log(p)/np.log(2))
        return s
    def gini(self,*args):
        s=1
        sum_args=sum(args)
        for arg in args:
            if arg==sum_args:
                return 0
            p=arg/sum_args
            if p==0:
                continue
            s-=p**2
        return s
    def impurity_advanced(self,data,var_target,var,impurity,print_table=False):
        data_new=np.array(data[[var,var_target]])
        data_new=pd.crosstab(*(data_new.T),rownames=[var],colnames=[var_target],margins=True,margins_name="Total")
        if print_table:
            print(data_new)
        data_new=np.array(data_new)
        S=0
        if var_target==var:
            return self.entropy(*data_new[-1,:][:-1])
        for i in range(data_new.shape[0]-1):
            if(impurity=="entropy"):
                S+=self.entropy(*data_new[i,:][:-1])*(data_new[i][-1]/data_new[-1,-1])
            if(impurity=="gini"):
                S+=self.gini(*data_new[i,:][:-1])*(data_new[i][-1]/data_new[-1,-1])
        return S
    def information_gain(self,data,var_target,var,impurity,S_whole=None):
        if S_whole==None:
            S_whole=self.impurity_advanced(data,var_target,var_target,impurity)
        gain=S_whole-self.impurity_advanced(data,var_target,var,impurity)
        return gain
    def max_information_gain(self,data,var_target,impurity):
        var_list=list(data.keys())
        var_list.remove(var_target)
        gain_list=[]
        S_whole=self.impurity_advanced(data,var_target,var_target,impurity)
        for var in var_list:
            gain_list.append(self.information_gain(data,var_target,var,impurity,S_whole))
        sorted_gain_list=sorted(gain_list)
        sorted_var_list=[]
        for i in range(len(gain_list)):
            sorted_var_list.append(var_list[gain_list.index(sorted_gain_list[i])])
       # print(sorted_var_list)
       # print(sorted_gain_list)
        return sorted_var_list[-1]
    def recursive_decision_tree(self,data,var_target,impurity,max_depth):
        if max_depth==0:
            return max(list(data[var_target]))
        data_list={}
        s_target=list(set(data[var_target]))
        if len(s_target)<=1:
            return s_target[0]
        node=self.max_information_gain(data,var_target,impurity)
        s_temp=list(set(data[node]))
        for i in s_temp:
            data_temp=data[data[node]==i].drop([node],axis=1)
            data_list[node+"->"+str(i)]=self.recursive_decision_tree(data_temp,var_target,impurity,max_depth-1)
        return data_list
    def fit(self,data,var_target):
        impurity=self.impurity
        max_depth=self.max_depth
        self.model=self.recursive_decision_tree(data,var_target,impurity,max_depth)
        return self.model
    def predict_helper(self,model,data,l):
        d_keys=list(model.keys())
        for i in range(len(d_keys)):
            data_split=data[data[d_keys[i].split("->")[0]]==d_keys[i].split("->")[1]]
            if type(model[d_keys[i]])==str:
                for j in list(data_split.index):
                    l[j]=model[d_keys[i]]
            else:
                self.predict_helper(model[d_keys[i]],data_split,l)
    def predict(self,data):
        l=[0 for i in range(data.shape[0])]
        self.predict_helper(self.model,data,l)
        return l
    def grapher_helper(self,model,graph):
        d_keys=list(model.keys())
        for i in range(len(d_keys)):
            l=d_keys[i].split("->")
            if type(model[d_keys[i]])==str:
                graph.node(model[d_keys[i]],color="green")
                graph.edge(l[0].replace("\n",""),model[d_keys[i]],label=l[1],constraint="True")
            else:
                graph.edge(l[0].replace("\n",""),list(model[d_keys[i]].keys())[0].split("->")[0].replace("\n",""),label=l[1],constraint="True")
                self.grapher_helper(model[d_keys[i]],graph)
    def grapher(self):
        model=self.model
        graph=graphviz.Digraph(name="Decision Tree",node_attr={'color': 'lightblue2', 'style': 'filled'},engine="dot")
        self.grapher_helper(model,graph)
        return graph
```

```python
def accuracy(y_pred,y_test):
    n=len(y_pred)
    acc=0
    for i in range(n):
        if y_pred[i]==y_test[i]:
            acc+=1
    return acc/n
```

```python
model=decision_tree(impurity="entropy",max_depth=30)
model.fit(data,var_target)
y_test=list(data[var_target])
y_pred=model.predict(data.drop([var_target],axis=1))
print("accuracy is",accuracy(y_pred,y_test))
model.grapher()
```
